```python
# Define input and output paths
restrepo_input_path = Path().resolve().parents[1] / "data" / "raw" / "features" / "FeaturesPhrampusLee_TOCSedRate_updated"
output_path = Path().resolve().parents[1] / "data" / "interim" / "worldfeatures" / "WorldFeaturesAllnoNaN_SedTOC"
```

```python
# Get Restrepo feature files
restrepo_files = get_restrepo_files(restrepo_input_path)
```

```python
feature_selection = pd.read_csv(Path().resolve().parents[1] / "data" / "interim" / "selectedFeatureLists" / "selectedfeatures_v_men.txt").values
# Load datasets
dataset, datasets, corrupted_files = load_datasets(restrepo_files, restrepo_input_path, feature_selection)
```

```python
chunks = 360  # Adjust the number of chunks as per your requirement
lat_step = int(dataset.shape[0]/chunks)
def process_chunk(ii):
    feature_chunk = np.zeros([int(dataset.shape[0] / chunks * dataset.shape[1]), len(datasets)])

    for idx in range(len(datasets)):
        if datasets[idx] is not np.nan:
            feature_chunk[:, idx] = np.asarray(datasets[idx][ii * lat_step:ii * lat_step + lat_step, :]).flatten(order="C")
        else:
            feature_chunk[:, idx] = np.nan

    save_name = "Sed_world_feature_" + str(ii).zfill(3)
    np.save(Path(output_path / save_name), feature_chunk)
    print("Chunk " + str(ii + 1) + " of " + str(chunks) + " written!")


pool = Pool(processes=64)  # Adjust the number of processes as per your requirement

pool.map(process_chunk, range(chunks))

pool.close()
pool.join()
```
